# Memory-Based To-Do List

```python
Lists = ['apple', "Banana"]

while True:
    print("\n----> MENU <----")
    print("1. View")
    print("2. Add")
    print("3. Remove by using name")
    print("4. Delete by index")
    print("5. Update")
    print("6. Exit")

# input your choice have an index and item name
    choice = input("Enter your choice: ")

# view list to check all of the item
    if choice == '1':
        print("----> LIST ITEM <----")
        for i, item in enumerate(Lists):
            print(f"Index {i} | Item name: {item}")
# add more item in the list
    elif choice == '2':
        item = input("Enter item: ")
        Lists.append(item)
# remove item by using key-word remove
    elif choice == '3':
        item = input("Enter item Name to delete: ")
        if item in Lists:
            Lists.remove(item)
            print(f"\nItem name {item} has been removed now.")
        else:
            print("\nThe item do not have in Lists checking again...")
# delete item by using index (key-word .pop())
    elif choice == '4':
        index_input = input("Enter index of Item list: ")

        if index_input.isdigit():
            idx = int(index_input)
            if idx < len(Lists):
                removed_item = Lists.pop(idx) 
                print(f"Index {idx} | Item name: {removed_item}, has been deleted now.")
            else:
                print(f"Error: Index {idx} is out of range!")
        else:
            print("Error: Please enter a valid number!")
# using to update an item or change item name 
    elif choice == '5': 
        index_input = input("Enter index to update: ")
        if index_input.isdigit():
            idx = int(index_input)
            if 0 <= idx < len(Lists):
                old_val = Lists[idx]
                new_val = input(f"Enter new value for '{old_val}': ")
                Lists[idx] = new_val
                print(f"Index {idx} updated successfully!")
            else:
                print("Index out of range!")
        else:
            print("Please enter a valid number!")
# Exit part
    elif choice == '6':
        print("Thank you, Have a nice day!")
        break
    else:
        print("Invalid choice, Try again...")
```

## List

```python
# it is a simple List 

# list using to store the data and allow to change, update, delete (mutable)
item = ["apply", "cake", 1.5 , 20.9]

item[0] = "Banana"         # use to update the data 
item.append("Computer")    # use to add more item
item.pop(2)                # use to delete by using index(pop)
item.remove(20.9)          # use to remove item by name

# we using this to print an see all what the list have
for i, obj in enumerate(item):   
            print(f"Index: {i} | Item name: {obj}")
```

## Dictionaries

```python
# this is talking about key:value in Dictionary

quiz_data = {
    "Cambodia": "Phnom Penh",
    "Japan": "Tokyo",
    "France": "Paris",
    "South Korea": "Seoul",
    "United Kingdom": "London",
    "China": "Beijing",
    "USA": "Washington D.C.",
    "Vietnam": "Hanoi",
    "Italy": "Rome",
    "Germany": "Berlin"
} 

score = 0
for country, capital in quiz_data.items():
    user_answer = input(f"What is the capital of {country}? ")
    
    if user_answer.strip().lower() == capital.lower():
        print("Correct!")
        score += 1  
    else:
        print(f"Wrong! The correct answer is {capital}.")

print("---------------------------------------")
print(f"Game Over! Your final score is: {score}/{len(quiz_data)}")
```

## Set: Unique Collections

```python
# set it is a Uniqe collection that we use to make the word have only one time 
# so if that "Name" have 2 time {set} fill to collect only one 

my_set = {"apple", "banana", "apple", "orange", "banana"}
print(my_set)

set_A = {1,2,3,4,5,6}
set_B = {5,6,7,8,9,10}

# using this build-in fuction to print all the number in one line
print(set_A | set_B)         
print(set_A.union(set_B))

# using to see which number have both in A and B

number = set_A & set_B
print(number)
print(set_A.intersection(set_B))

list_Number = set_A | set_B
print(min(list_Number))
print(max(list_Number))

```
