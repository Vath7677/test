#The Basic of guessing game

```python
import random

# excersice point 1

# this use to random a number form 1 to 100
secret_number = random.randint(1,100)

guess = int(input("Enter guessing number(1-100): "))

# checking the condition it 's much to your answer or not
if guess == secret_number:
    print("You are right!")
elif guess > secret_number:
    print("Too high...! The number is",secret_number)
else:
    print("Too low...! The number is",secret_number)
```

```python
import random  

# excersice point 2 & 3 & 4

# this is an option to choice
options = ["rock", "paper", "scissors"]
playing = True

userScore = 0
computerScore = 0

print("--- Welcome to Rock, Paper, Scissors! ---")

while playing:  
    # input your choice
    user_choice = input("\nChoose Rock, Paper, or Scissors (or 'q' to quit): ").lower()

    # if u want to quite input q 
    if user_choice == 'q':
        playing = False
        break

    if user_choice not in options:
        print("Invalid choice, try again.")
        continue

        # this is computer random choice
    computer_choice = random.choice(options)
    print(f"Computer chose: {computer_choice}")
    
    if user_choice == computer_choice:
        print("It's a tie!")
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win!")
        userScore += 1
    else:
        print("The computer wins!")
        # make a counter point of computer
        computerScore += 1

    print(f"Your score: {userScore} | Computer score: {computerScore}")
print("Thanks for playing!")
```
