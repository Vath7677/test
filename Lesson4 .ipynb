# Function

```python
#### function is a block of code that do an action that we have set for it 

# put this as an function to call to use and it is make our code clean 

# using function to return agument by parameter 
def c_to_f(c): 
    return (c * 9/5) + 32
def m_to_k(m): 
    return m * 1.60934
def g_to_K(g):
    return g / 1000 

def Converter():
    while True:
        print("--------------------------")
        print("1. Celsius to Fahrenheit")
        print("2. Miles to Kilometers")
        print("3. Grams to Kilograms")
        print("4. Exit")

        choice = input("Select an option(1-4): ")

        if choice == '1':
            c = float(input("Enter Celsius: "))
            result = c_to_f(c)
            print(f"Celsius: {c} toFahrenheit: {result}")
        elif choice == '2':
            m = float(input("Enter Mile: "))
            result = m_to_k(m)
            print(f"Mile: {m} to Km {result}")
        elif choice == '3':
            g = float(input("Enter Gram: "))
            result = g_to_K(g)
            print(f"Gram: {g} to Kilogram: {result}")
        elif choice == '4':
            print("Bye bye...!")
            break
        else:
            print("Ivalid option...!")
            
# call the function to print
Converter()
            
```

## Generator password

```python
%%writefile generator.py   # this one is like we save file name in one file 
                    
# use in the web broswer because i cannot save the file and put the name so i can import to use

import string
import random

def generate_password(length=12, use_numbers=True, use_symbols=True):
    # it is use to store a number or letter that we all have
    pool = string.ascii_letters  # or we can use "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ" 

    # this mean if the use_number is true it will be use pool(Letter + number .digits)
    if use_numbers:
        pool += string.digits    

    #if the use_symbol is true it will be use pool(Letter + number {.digits} + symbol { .punctuation} )
    if use_symbols:
        pool += string.punctuation 
    
    # Generate the password by using rondom and combind it together 
    password = "".join(random.choice(pool) for i in range(length))
    
    return password
```

```python
# we use this to import the code from other file to use 
from generator import generate_password   # Modules and Package

# generate_password this one we import form function of the first file

def generator_password_menu():
    while True:
        print("\n--- Password Tool ---")
        print("1. Generate password")
        print("2. Exit")
        choice = input("Select an option: ")

        if choice == '1':
            try:
                size = int(input("How many characters? "))
                nums = input("Include numbers? (y/n): ").lower() == 'y'
                syms = input("Include symbols? (y/n): ").lower() == 'y'
                
                # give an data to Funtion to use in other file
                new_pass = generate_password(length=size, use_numbers=nums, use_symbols=syms)
                print(f"\nYour secure password is: {new_pass}")
            except ValueError:
                print("Error: Please enter a number for the length!")
        elif choice == '2':
            print("Exit the tool password create...")
            break
        else:
            print("Invalid choice!")

generator_password_menu()
```
